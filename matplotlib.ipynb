```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
x =  np.linspace(0,5,11)
x
```

```python
y =  x**2
y
```

```python
plt.plot(x,y)
plt.title('line graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

SUBPLOTS

```python
plt.subplot(2,2,1)
plt.plot(x,y**2)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,4)
plt.plot(x,y)
```

object oriented way

```python
fig =  plt.figure(figsize=(20,8))
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.4,0.5,.3,.3])
axis2.plot(x,y)
```

TYPES OF PLOT

```python
plt.scatter(x,y)
```

```python
plt.hist(x)
```

plt.boxplot(x)

SAVING YOUR PLOT 

```python
fig = plt.figure()
axis3 = fig.add_axes([0.1,0.1,0.8,0.8])
axis3.plot(x,y)
plt.savefig("basicplot.png")
```
